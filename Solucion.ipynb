```python
import numpy as np
from src import gauss_jacobi
from src import gauss_seidel
import matplotlib.pyplot as plt

# Sistema de ecuaciones
A = np.array([[2, 10], [3, 2]], dtype=float)
b = np.array([16, 11], dtype=float).reshape(-1, 1)
x0 = np.array([1, 1], dtype=float).reshape(-1, 1)  # Aproximación inicial
tol = 1e-5  # Tolerancia
max_iter = 100  # Número máximo de iteraciones

x_jacobi, tray_jacobi = gauss_jacobi(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)
print("Método de Jacobi:")
print("Solución aproximada:", x_jacobi.T)

x_seidel, tray_seidel = gauss_seidel(A=A, b=b, x0=x0, tol=tol, max_iter=max_iter)
print("\nMétodo de Gauss-Seidel:")
print("Solución aproximada:", x_seidel.T)

tray_jacobi = np.array(tray_jacobi).squeeze()
tray_seidel = np.array(tray_seidel).squeeze() 

# Solución real para referencia
x_real = [3, 1]

plt.figure(figsize=(10, 6))

plt.plot(
    tray_jacobi[:, 0], tray_jacobi[:, 1], 
    marker="o", label="Jacobi", linestyle="--"
)

plt.plot(
    tray_seidel[:, 0], tray_seidel[:, 1], 
    marker="x", label="Gauss-Seidel", linestyle="-."
)

plt.scatter(x_real[0], x_real[1], color="red", label="Solución real", zorder=5)

plt.xlabel("$X_1$")
plt.ylabel("$X_2$")
plt.title("Trayectoria de las iteraciones: Métodos de Jacobi y Gauss-Seidel")
plt.axhline(0, color="gray", linewidth=0.5, linestyle="--")
plt.axvline(0, color="gray", linewidth=0.5, linestyle="--")
plt.legend()
plt.grid()

plt.show()
```
